```python
import numpy as np
import matplotlib.pyplot as plt
# from sklearn.metrics import f1_score
```

```python
def precision(tp, fp):
    # Make ver confident predictions
    return tp / (tp + fp)

def recall(tp, fn):
    # Capture as many positives as possible
    return tp / (tp + fn)

def f1(pres, rec):
    #if pres == 0 or rec == 0:
    #    return 0
    #else:
    return 2 * pres * rec / (pres + rec)

def fb(prec, rec, beta):
    return (1 + beta**2) * prec * rec / (beta**2 * prec + rec)
```

```python
# Create list with precision/recall values
res = 1 / 100
y_values = np.arange(0.01, 1 + res, res)
x_values = np.arange(0.01, 1 + res, res)
x_values = x_values[::-1]

# Calculate F1 score for each combination
f1_scores = []
for y in y_values:
    row = []
    for x in x_values:
        row.append(f1(y, x))
    f1_scores.append(row)

#np.reshape(f1_scores, (11,11))
```

```python
# Plot
ticks = [0, 0.2, 0.4, 0.6, 0.8, 1]

ax = plt.subplot(111)
im = ax.imshow(f1_scores)
plt.colorbar(im)

plt.xticks(np.arange(0, len(f1_scores), 20), ticks[::-1])
plt.yticks(np.arange(0, len(f1_scores), 20), ticks)
ax.set_xlabel('Recall')
ax.set_ylabel('Precision')
ax.set_title('F1 score')
```

```python
X, Y = np.meshgrid(x_values, y_values)
Z = f1(X, Y)

plt.contour(X, Y, Z)
plt.colorbar()
```
